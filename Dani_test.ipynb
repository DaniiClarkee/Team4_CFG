```python
import pandas as pd
df_raw = pd.read_csv("data.csv", index_col=[0])
```

```python
df_raw.head()
```

```python
df_raw['diagnosis'].value_counts()
```

```python
col_names = df_raw.columns
num_data = df_raw.shape[0]
for c in col_names:
    num_non = df_raw[c].isin([" ?"]).sum()
    if num_non > 0:
        print (c)
        print (num_non)
        print ("{0:.2f}%".format(float(num_non) / num_data * 100))
        print ("\n")
```

```python
df_raw = df_raw[df_raw["diagnosis"] != " ?"]
```
